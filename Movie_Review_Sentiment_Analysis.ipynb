```python
import numpy as np
import pandas as pd
```

Data needs to be cleaned by removing html tags, special characters
#Steps:
1. Sample 10000 rows
2. Remove html tags
3. Remove special characters
4. Convert text to lower case
5. Remove Stop words: Stop words are common words in a language that do not add much meaning to text analysis. like "is," "the," "a," "an," "in," "at," "for," "of," "on," "and," "with," etc.
6. Stemming : text preprocessing technique used in Natural Language Processing (NLP), including sentiment analysis, to reduce words to their root or base form. e.g. "Running" â†’ "Run"

```python
df = df.sample(10000)
df.shape
```

```python
df.info()
```

```python
df['sentiment'].replace({'positive':1, 'negative':0}, inplace=True)
```

```python
df.head()
```

```python
import re
clean = re.compile('<.*?>')
re.sub(clean, '', df.iloc[2].review)
```

```python
#function to clean html tags
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['review']=df['review'].apply(clean_html)
```

```python
df['review'].head()
```

```python
# converting review to lower text

def convert_lower(text):
  return text.lower()
```

```python
df['review'] = df['review'].apply(convert_lower)
```

```python
# function to remove special characters

def remove_special(text):
  x=''

  for i in text:
    if i.isalnum():
      x=x+i
    else:
      x=x + ' '
  return x
```

```python
df['review']=df['review'].apply(remove_special)
```

```python
# Remove Stop words

import nltk
nltk.download('stopwords')
#natural language toolkit
```

```python
from nltk.corpus import stopwords
```

```python
print('Total number of stopwords in english =',len(stopwords.words('english')))

print(stopwords.words('english'))
```

```python
df
```

```python
def remove_stopwords(text):
  x=[]
  for i in text.split():

    if i not in stopwords.words('english'):
      x.append(i)
  y=x[:]
  x.clear()
  return y
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
df
```

```python
# Stemming

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
y=[]

def stem_words(text):
  for i in text:
    y.append(ps.stem(i))
  z=y[:]
  y.clear()
  return z
```

```python
df['review']=df['review'].apply(stem_words)
```

```python
df
```

```python
# join back

def join_back(list_input):
  return " ".join(list_input)
```

```python
df['review']=df['review'].apply(join_back)
```

```python
df['review']
```

```python
X = df.iloc[:,0:1].values
```

```python
X.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000)
#max_features=500
```

```python
X= cv.fit_transform(df['review']).toarray()
```

```python
X.shape
```

```python
y = df.iloc[:, -1].values

# these are the labels capturing only last column
```

```python
y.shape
```

split the data into 2 parts
* 8000 rows : Training set
* 2000 rows : Test set
* Test set results are already known

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
```

```python
y_pred1=clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
y_pred3=clf3.predict(X_test)
```

```python
y_pred1.shape
```

```python
#calculate accuracy

from sklearn.metrics import accuracy_score
```

```python
print('Gaussian', accuracy_score(y_test, y_pred1))
print('Multinomial', accuracy_score(y_test, y_pred2))
print('Bernaulli', accuracy_score(y_test, y_pred3))
```

After performing sentiment analysis on a sample of 10000 from a dataset of 50,000 movie reviews, applying NaÃ¯ve Bayes classifiers after text preprocessing. The accuracy scores with max_features=1000 are:
* Gaussian 0.79
* Multinomial 0.829
* Bernaulli 0.822

#Insights & Interpretation
1. Why is MultinomialNB the Best?

* âœ… Best for text classification â€“ It assumes word frequency follows a multinomial distribution, making it ideal for Bag-of-Words (BoW) or TF-IDF representations.
* âœ… Captures word importance â€“ Since sentiment is often determined by the presence of positive or negative words, MultinomialNB is well-suited for detecting sentiment patterns.

2. Why Did BernoulliNB Perform Similarly?
* âœ… BernoulliNB works well with binary features (word presence/absence).
* âœ… Since you removed stop words and limited vocabulary (max_features=1000), BernoulliNB was effective in sentiment classification.
* âœ… However, it ignores word frequency, which is why MultinomialNB slightly outperformed it.

3. Why Did GaussianNB Perform the Worst?
* âŒ Not suited for text classification â€“ GaussianNB assumes continuous numeric features (like heights, weights, etc.), whereas text data is categorical.
* âŒ Less effective with word frequency â€“ Since word frequencies drive sentiment, GaussianNBâ€™s assumption of a normal distribution does not align well with text data.

# Key Takeaways
âœ… MultinomialNB is the best choice for sentiment analysis since it effectively models word frequency.

âœ… BernoulliNB is also effective, especially when using binary presence/absence features.

âœ… GaussianNB is unsuitable for text data and performs worse than both alternatives.
